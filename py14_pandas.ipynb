## indexing, slicing

* `iloc`: integer location. 정수 인덱스(ndarray의 기본 인덱스) 기반으로 원소 참조 또는 슬라이싱.
* `loc`: location. 라벨 기반의 원소 참조 또는 슬라이싱.

```python
np.random.seed(1)
s1 = pd.Series(data = np.random.randint(-10, 10, size = 6))
s1
```

```python
# iloc를 사용한 인덱싱
print(s1.iloc[0])       # Series에서 첫번째 원소
print(s1.iloc[-1])      # Series에서 마지막 원소
```

```python
# iloc를 사용한 슬라이싱
print(s1.iloc[:3])      # Series에서 첫 3개 원소 자르기
print(s1.iloc[-3:])     # Series에서 마지막 3개 원소 자르기
```

```python
# loc를 사용한 인덱싱
print(s1.loc[0])    # Series의 첫번째 원소
# print(s1.loc[-1])   # Series의 마지막 원소 -> 음수 인덱스를 사용하지 못 함. label로 붙여진대로만 사용해야 함.
print(s1.loc[5])
```

```python
# loc를 사용한 슬라이싱
print(s1.loc[:2])   # Series에서 첫 3개 원소 자르기
```

```python
np.random.seed(42)
s2 = pd.Series(data = np.random.randint(-10, 11, size = 6), index = ['a', 'b', 'c', 'd', 'e', 'f'])
s2
```

```python
print(s2.iloc[0])
print(s2.iloc[-1])
print(s2.iloc[-3:])
```

```python
print(s2.loc['a'])
print(s2.loc['f'])
print(s2.loc[:'c'])
```

# DataFrame

* 데이터를 2차원(행 row, 열 column) 형태로 저장하는 데이터 타입(클래스).
* 데이터베이스의 테이블과 비슷한 구조.
* DataFrame에서 컬럼은 Series 타입.

```python
# dict
data = {
    'city' : ['서울'] * 3 + ['경기'] * 3,
    'year' : [2021, 2022, 2023] * 2,
    'pop' : np.random.random(size = 6)
}
data
```

```python
df = pd.DataFrame(data)
```

```python
print(df)       # __str__
```

```python
df      # __repr__
```

## DataFrame에서 조건에 맞는 행을 선택 - boolean indexing

```python
df
```

```python
# select * from df where year = 2023;
df.year == 2023
```

```python
df[df.year == 2023]
```

* boolean indexing을 사용할 때는, 파이썬의 논리 연산자(`and, or, not`)를 사용할 수 없음!
* boolean indexing에서는 `&, |, ~` 연산자를 사용함!
* 조건식과 조건식은 `()`를 사용해서 구분, 연산의 순서를 명시해야 함.

```python
# city가 '서울'이거나, pop이 0.65 이하인 행들을 선택:
df[(df['city'] == '서울') | (df['pop'] <= 0.65)]
```

```python
# 경기도의 2023년 자료를 선택
df[(df.city == '경기') & (df.year == 2023)]
```

## DataFrame에서 행과 열을 함께 선택

```python
df[df.city == '서울'][['year', 'pop']]
```

```python
df.loc[df.city == '서울', ['year', 'pop']]  # df.loc[row, column]
```

# DataFrame 연습

```python
file_path = 'https://raw.githubusercontent.com/JakeOh/20230228_itwill_java140_lab_python/main/csv_exam.csv'
```

```python
# github에 저장된 CSV 파일을 읽어서 데이터 프레임 생성:
exam = pd.read_csv(file_path)
exam
```

* 데이터프레임의 첫 5개 행을 출력
* 데이터프레임의 마지막 5개 행을 출력
* 숫자 타입 컬럼들의 기술 통계량을 출력
* class 컬럼의 빈도수를 찾음
* 수학, 영어, 과학 컬럼의 기술 통계량을 출력
* 1반 학생들의 데이터를 출력
* 수학 평균을 찾고 출력
* 수학 점수가 평균 이상인 학생들의 데이터를 출력
* 1반 학생들의 수학 점수 평균을 찾고 출력
* 2반 학생들의 수학 점수 평균을 찾고 출력
* 반(class)별 수학 점수 평균을 찾고 출력
* 세 과목의 점수가 모두 평균 이상인 학생들의 데이터를 출력

```python
# 데이터프레임의 첫 5개 행을 출력
exam.head()
```

```python
# 데이터프레임의 마지막 5개 행을 출력
exam.tail()
```

```python
# 숫자 타입 컬럼들의 기술 통계량을 출력
exam.describe()
```

```python
# class 컬럼의 빈도수를 찾음
exam['class'].value_counts()
```

```python
# 수학, 영어, 과학 컬럼의 기술 통계량을 출력
exam[['math', 'english', 'science']].describe()
```

```python
# 1반 학생들의 데이터를 출력
exam[exam['class'] == 1]
```

```python
# 수학 평균을 찾고 출력
print(exam['math'].mean())
print(exam['math'].std())
```

```python
# 수학 점수가 평균 이상인 학생들의 데이터를 출력
exam[exam.math >= exam.math.mean()]
```

```python
# 1반 학생들의 수학 점수 평균을 찾고 출력
exam[exam['class'] == 1].math.mean()
```

```python
# 2반 학생들의 수학 점수 평균을 찾고 출력
exam[exam['class'] == 2]['math'].mean()
```

```python
# 반(class)별 수학 점수 평균을 찾고 출력
for x in range(1, len(exam['class'].value_counts()) + 1):
    print(f'{x}반 수학 점수 평균:', exam[exam['class'] == x]['math'].mean())
```

```python
exam.groupby('class')['math'].mean()
```

```python
for x in range(1, 6):
    math_mean = exam[exam['class'] == x].math.mean()
    print(f'class {x} : {math_mean}')
```

```python
exam_dict = {'class' : np.arange(1, 6)}
means = []
for x in range(1, 6):
    math_mean = exam[exam['class'] == x].math.mean()
    means.append(math_mean)

print(means)

exam_dict = {
    'class': np.arange(1, 6),
    'math_mean': means
}

print(exam_dict)

df = pd.DataFrame(exam_dict)
print(df)
```

```python
# 세 과목의 점수가 모두 평균 이상인 학생들의 데이터를 출력
exam[
    (exam['math'] >= exam['math'].mean()) &
    (exam['english'] >= exam['english'].mean()) &
    (exam['science'] >= exam['science'].mean())
]
```

```python
# 1반 학생들의 수학 평균 점수 이상인 학생들
exam[exam.math >= exam[exam['class'] == 2].math.mean()]
```
