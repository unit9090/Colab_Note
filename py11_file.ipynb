# 파일 쓰기(wrtie)

파일 쓰기 순서:
1. 파일 열기(open)
2. 파일 내용 쓰기(write)
3. 파일 닫기(close)

`open(file, mode='rt', ...)`:
* file: 폴더(디렉토리)를 포함한 파일 이름.
* mode: 파일을 여는 방법(모드).
    * r(read): 읽기. 기본값.
        * 파일이 없으면 FileNotFoundError가 발생.
    * w(write): 쓰기.
        * 파일이 있으면 기존 내용이 모두 삭제되고 파일의 처음부터 다시 쓰기 시작.
        * 파일이 없으면 새로운 파일을 생성하고 파일 쓰기가 됨.
    * a(append): 이어쓰기. 기존 작성된 내용 이후부터 이어서 쓰기.
        * append 모드도 write 모드와 같은 기능을 한다.
        * 파일이 있으면 기존 파일 뒤에 이어쓰기
        * 파일이 없으면 새로운 파일을 생성하고 이어쓰기
    * t(text): 텍스트 (읽기/쓰기). 인코딩된 문자열.
    * b(binary): 이진 데이터(읽기/쓰기). (예) 이미지, 동영상, ...

```python
# 1. 파일을 텍스트 쓰기 모드로 open:
f = open('test1.txt', mode='wt')

# 2. 파일 쓰기:
f.write('안녕하세요\n')
f.write('Hello, python!\n음....집갈랭....\n나도 갈 수 있어!!!\n')
f.write('1234567890\n')

# 3. 파일 닫기:
f.close()
```

```python
# with ~ as 구문: 오픈된 리소스(파일)에서 close() 메서드를 자동으로 호출.
# as 뒤에 나오는 것이 변수의 별명
with open('test2.txt', 'wt') as f:
    f.write('hello\n')
    f.write('즐거운 월요일\n')
```

# 파일 읽기

```python
with open('test1.txt', mode='rt') as f:
    # readline: 줄바꿈이 있는 곳까지 읽음. 대신, 줄바꿈이 몇 번 일어나는지 알아야 함.
    #           또 줄바꿈까지 출력하기 때문에 이상하게 나올 수 있음.
    # 문자열.strip(): 문자열의 왼쪽과 오른쪽에 있는 공백 문자들(space, tab, 줄바꿈)을 제거.
    line = f.readline()
    print(line.strip())
    line = f.readline()
    print(line.strip())
    line = f.readline()
    print(line.strip())
```

```python
with open('test2.txt', mode = 'rt') as f:
    for line in f:
        # 파일의 처음부터 끝까지 readline()을 호출
        print(line.strip())
```
