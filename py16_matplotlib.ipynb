# `matplotlib.pyplot` 모듈을 사용한 시각화

1. `figure` 객체 생성.
    * `figure = plt.figure()`
2. `figure` 객체에 subplot(axis)을 생성해서 추가.
    * figure.add_subplot(행의 개수(nrows), 열의 개수(ncols), 인덱스(* subplot의 인덱스는 1부터 시작))
    * `axis = figure.add_subplot(111)`
    * `axis = figure.add_subplot(1, 1, 1)`
3. subplot에 그래프 그림.
4. 그래프의 여러가지 옵션들을 설정.
5. 그래프 보여주기.

```python
# 1. figure 객체 생성
figure = plt.figure()

# figure 객체에 subplot(axis)을 생성해서 추가.
# axis = figure.add_subplot(1, 1, 1)
axis = figure.add_subplot(111)

# 행을 나눈 경우
# axis = figure.add_subplot(211)
# figure.add_subplot(212)

# 3. subplot에 그래프 종류를 선택해서 그림
axis.hist(x = tips['total_bill'])

# 4. 그래프의 옵션 설정
axis.set_xlabel('Total Bill')
axis.set_ylabel('Count')
axis.set_title('Histogram')
axis.grid()

# 5. 그래프 보여주기
plt.show()
```

figure에 2개의 subplot을 상하로 배치하고 그래프 그리기

```python
# figure 생성
# 표의 크기 설정: figsize = (가로 길이, 세로 길이)
fig = plt.figure(figsize = (6, 8))

# 2. subplot을 figure에 추가
ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)

# 3. 그래프
ax1.hist(x = tips.total_bill)
ax2.scatter(x = tips.total_bill, y = tips.tip)

# 4. 옵션
ax1.grid()
ax2.grid()

ax1.set_xlabel('Total Bill')
ax1.set_ylabel('Count')
ax2.set_xlabel('Total Bill')
ax2.set_ylabel('Tip')

# 5. 보여주기
plt.show()
```

figure에 2개의 subplot 좌우로 배치하고 hist, scatter 그래프를 그리기

```python
# figure 생성
# 표의 크기 설정: figsize = (가로 길이, 세로 길이)
fig = plt.figure(figsize = (12, 4))

# 2. subplot을 figure에 추가
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

# 3. 그래프
ax1.hist(x = tips.tip, color='black')
ax2.scatter(x = tips.total_bill, y = tips.tip, color='black')

# 4. 옵션
ax1.grid()
ax1.set_xlabel('Tip')
ax1.set_ylabel('Count')
ax1.set_title('Histogram')

ax2.grid()
ax2.set_xlabel('Total Bill')
ax2.set_ylabel('Tip')
ax2.set_title('Scaatter plot')

# 5. 보여주기
plt.show()
```

# `pyplot` 모듈의 top-level 함수

## Histogram

* 연속형 변수를 구간으로 나눠서, 그 구간 안에 포함된 자료의 개수를 막대의 높이로 표현한 그래프.
* 연속형 변수의 분포를 알 수 있음.

hist 함수 파라미터:

* x: x축에 사용할 변수 이름. 히스토그램을 그리기 위한 자료.
* bins: 막대의 개수, 구간
* color: 막대 안쪽을 채우는 색깔
* edgecolor: 막대 테두리 색깔
* density: False - 구간 안에 포함된 개수, True - 밀도(비율)

```python
# 1. figure 생성 + 2. subplot 1개 생성. + 3. histogram 그림.
plt.hist(x = tips.total_bill, color='lightgray', edgecolor = 'black', bins = 10, density = True)

# 4. 옵션 설정
plt.xlabel('Total bill')
plt.ylabel('Count')
plt.title('Histogram')
plt.grid()

# 5. 보여주기
plt.show()
```

## 산점도(Scatter plot)

두 변수(컬럼) 사이의 상관 관계를 파악하기 위해서 사용.

```python
# tip ~ total_bill 상관 관계
# y축 변수 ~ x축 변수
plt.scatter(
    x = tips.total_bill,
    y = tips.tip,
    edgecolors= 'black',
    alpha = 0.3,        # alpha: 0(투명) ~ 1.0(불투명)
    s = 100,            # s: size. 마커(점)의 크기
    c = 'red',          # c: color. 마커(점)의 색깔
    marker = 's'        # marker: 마커의 종류(모양). 'o', '^', 'v', '*', 's', 'h', ...
)

plt.grid()
plt.xlabel('Total bill')
plt.ylabel('Tip')

plt.show()
```

```python
# tip ~ size 상관 관계
plt.scatter(
    x = tips['size'],
    y = tips.tip,
    edgecolors= 'black',
    alpha = 0.3,
    s = 100,
    c = 'gray',
    marker = 'h'
)

plt.grid()
plt.xlabel('Size')
plt.ylabel('Tip')

plt.show()
```

## 선 그래프(Line plot)

주로 시계열 데이터(시간에 따라서 변하는 데이터: 주가, 환율, 날씨, ..)를 시각화.

```python
np.random.seed(1)   # 난수 생성하는 순서.
df = pd.DataFrame(data = {
    'time': np.arange(1, 11),       # [1, 2, 3, ..., 9, 10]
    'price': np.random.rand(10)     # [0, 1.0) 범위의 난수 10개를 갖는 1차원 배열
})

df
```

```python
# price ~ time
plt.plot(
    df.time,        # x축 데이터
    df.price,       # y축 데이터
    'm*:'
    # format 문자열(3번째 arg): marker의 색상, marker의 모양, 선의 종류(-, --, :)
    # 색상: b(lue), r(ed), g(reen), (blac)k, y(ellow), c(민트색?), m(보라? 핑크?), w(hite)...
    # 마커 종류: o, ^, v, *, s, +, ...
    # 선 종류: -(solid line), :(dotted line), --(dashed line), ...
)

plt.grid()
plt.xlabel('Time')
plt.ylabel('Price')

plt.show()
```

```python
# variable-length keyword argument(**kwargs) 사용방법:
plt.plot(
    df.time,
    df.price,
    c = 'crimson',      # color
    marker = '^',       # marker
    linestyle = 'dotted'    # line style
)

plt.show()
```

1개의 subplot에 선 그래프 3개를 중첩시켜서 표현.

```python
np.random.seed(1)
df = pd.DataFrame(data = {
    'time': np.arange(1, 11),
    'price1': np.random.rand(10),
    'price2': np.random.rand(10),
    'price3': np.random.rand(10)
})
df
```

```python
# price1 ~ time
plt.plot(
    df.time,
    df.price1,
    'bo-',
    label = 'price1'
)

# price2 ~ time
plt.plot(
    df.time,
    df.price2,
    'rs--',
    label = 'price2'
)

# price3 ~ time
plt.plot(
    df.time,
    df.price3,
    'g+:',
    label = 'price3'
)

plt.grid()
plt.legend()        # 범례(legend) 표시

plt.show()
```

# 연습문제

## Ex 1.

*  4개의 subplot을 2x2 모양으로 배치.
*  4개의 subplot에 각각 성별 빈도수, 흡연여부 빈도수, 요일 빈도수, 시간 빈도수 그래프를 그리세요.

```python
# bar plot을 그리기 위한 데이터
gender_count = tips.sex.value_counts()
smoker_count = tips.smoker.value_counts()
day_count = tips.day.value_counts()
time_count = tips.time.value_counts()
```

```python
# 2 x 2 모양의 subplot들을 생성
fig, ax = plt.subplots(ncols = 2, nrows = 2, figsize = (8, 8))

ax[0, 0].bar(x = gender_count.index, height = gender_count.values)
ax[0][1].bar(x = smoker_count.index, height = smoker_count.values)
ax[1][0].bar(x = day_count.index, height = day_count.values)
ax[1][1].bar(x = time_count.index, height = time_count.values)

for i in range(2):
    for j in range(2):
        ax[i, j].grid()

plt.show()
```

```python
categories = ['sex', 'smoker', 'day', 'time']
counts = [tips[x].value_counts() for x in categories]

fig, ax = plt.subplots(ncols = 2, nrows = 2, figsize = (8, 8))

num = 0
for i in range(2):
    for j in range(2):
        # index = i * 2 + j
        ax[i, j].bar(x = counts[num].index, height = counts[num].values)
        ax[i, j].grid()
        ax[i, j].set_title(categories[num])
        num += 1

plt.show()
```

## Ex 3.

*  `tip ~ total_bill`의 상관관계를 scatter plot으로 그리세요.
*  요일별로 점의 색깔, 모양을 다르게 표현하세요.
*  범례를 표시.

```python
days = tips.day.unique()
shape = ['o', 'x', 's', '^']
i = 0

for x in days:
    plt.scatter(
        x = tips[tips.day == x].total_bill,
        y = tips[tips.day == x].tip,
        alpha = 0.3,
        s = 30,
        marker = shape[i],
        label = x
    )
    i += 1

plt.grid()
plt.legend()
plt.xlabel('Total bill')
plt.ylabel('Tip')


plt.show()
```
