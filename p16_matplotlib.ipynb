# `matplotlib.pyplot` 모듈을 사용한 시각화

1. `figure` 객체 생성.
    * `figure = plt.figure()`
2. `figure` 객체에 subplot(axis)을 생성해서 추가.
    * figure.add_subplot(행의 개수(nrows), 열의 개수(ncols), 인덱스(* subplot의 인덱스는 1부터 시작))
    * `axis = figure.add_subplot(111)`
    * `axis = figure.add_subplot(1, 1, 1)`
3. subplot에 그래프 그림.
4. 그래프의 여러가지 옵션들을 설정.
5. 그래프 보여주기.

```python
# 1. figure 객체 생성
figure = plt.figure()

# figure 객체에 subplot(axis)을 생성해서 추가.
# axis = figure.add_subplot(1, 1, 1)
axis = figure.add_subplot(111)

# 행을 나눈 경우
# axis = figure.add_subplot(211)
# figure.add_subplot(212)

# 3. subplot에 그래프 종류를 선택해서 그림
axis.hist(x = tips['total_bill'])

# 4. 그래프의 옵션 설정
axis.set_xlabel('Total Bill')
axis.set_ylabel('Count')
axis.set_title('Histogram')
axis.grid()

# 5. 그래프 보여주기
plt.show()
```

figure에 2개의 subplot을 상하로 배치하고 그래프 그리기

```python
# figure 생성
# 표의 크기 설정: figsize = (가로 길이, 세로 길이)
fig = plt.figure(figsize = (6, 8))

# 2. subplot을 figure에 추가
ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)

# 3. 그래프
ax1.hist(x = tips.total_bill)
ax2.scatter(x = tips.total_bill, y = tips.tip)

# 4. 옵션
ax1.grid()
ax2.grid()

ax1.set_xlabel('Total Bill')
ax1.set_ylabel('Count')
ax2.set_xlabel('Total Bill')
ax2.set_ylabel('Tip')

# 5. 보여주기
plt.show()
```

figure에 2개의 subplot 좌우로 배치하고 hist, scatter 그래프를 그리기

```python
# figure 생성
# 표의 크기 설정: figsize = (가로 길이, 세로 길이)
fig = plt.figure(figsize = (12, 4))

# 2. subplot을 figure에 추가
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

# 3. 그래프
ax1.hist(x = tips.tip, color='black')
ax2.scatter(x = tips.total_bill, y = tips.tip, color='black')

# 4. 옵션
ax1.grid()
ax1.set_xlabel('Tip')
ax1.set_ylabel('Count')
ax1.set_title('Histogram')

ax2.grid()
ax2.set_xlabel('Total Bill')
ax2.set_ylabel('Tip')
ax2.set_title('Scaatter plot')

# 5. 보여주기
plt.show()
```

# `plt.subplots()` 함수

figure 생성 + subplot 추가

## figure에 1개의 subplot이 있는 경우

```python
# figure 생성, figure에 suplot을 추가.
fig, ax = plt.subplots()

# subplot에 그래프 그림.
ax.hist(x = tips.total_bill)

# 옵션

# 그래프 보여주기
plt.show()
```

## subplot 2개를 상하로 배치

```python
fig, ax = plt.subplots(nrows = 2, figsize = (6, 8))

# print(ax)   # subplot이 2개 이상인 경우 리턴 값 ax는 subplot들의 배열(ndarray)

ax[0].hist(x = tips.total_bill)
ax[1].scatter(x = tips.total_bill, y = tips.tip)

plt.show()
```

## 2개의 subplot을 좌우로 배치

```python
fig, ax = plt.subplots(ncols = 2, figsize = (12, 4))

# print(ax)   # subplot이 2개 이상인 경우 리턴 값 ax는 subplot들의 배열(ndarray)

ax[0].hist(x = tips.total_bill)
ax[1].scatter(x = tips.total_bill, y = tips.tip)

plt.show()
```

# `pyplot` 모듈의 top-level 함수

figure 생성 + subplot 1개 생성 + 그래프 종류

## Histogram

* 연속형 변수를 구간으로 나눠서, 그 구간 안에 포함된 자료의 개수를 막대의 높이로 표현한 그래프.
* 연속형 변수의 분포를 알 수 있음.

hist 함수 파라미터:

* x: x축에 사용할 변수 이름. 히스토그램을 그리기 위한 자료.
* bins: 막대의 개수, 구간
* color: 막대 안쪽을 채우는 색깔
* edgecolor: 막대 테두리 색깔
* density: False - 구간 안에 포함된 개수, True - 밀도(비율)

```python
# 1. figure 생성 + 2. subplot 1개 생성. + 3. histogram 그림.
plt.hist(x = tips.total_bill, color='lightgray', edgecolor = 'black', bins = 10, density = True)

# 4. 옵션 설정
plt.xlabel('Total bill')
plt.ylabel('Count')
plt.title('Histogram')
plt.grid()

# 5. 보여주기
plt.show()
```

## Box plot(상자 그림)

숫자 타입 변수(컬럼)의 기술 통계량(최솟값, 25%, 50% 75%, 최댓값)을 상자 모양으로 시각화.

```python
tips.total_bill.describe()
```

```python
plt.boxplot(x = tips.total_bill)

plt.grid()
plt.ylabel('Total bill')

plt.show()
```

```python
tips.tip.describe()
tips.tip
```

```python
plt.boxplot(x = tips.tip, vert = True)

plt.grid()
plt.xlabel('Tip')

plt.show()
```

## Bar plot(막대 그래프)

카테고리 타입 변수의 빈도수를 막대로 표현한 그래프.

```python
gender_count = tips.sex.value_counts()
gender_count
```

```python
type(gender_count)
```

```python
gender_count.index
```

```python
gender_count.values
```

```python
# 세로 막대 그래프(vertical bar plot)
# x: 인덱스, heigth: 막대의 높이.
# 빈도수 width: 막대 너비
plt.bar(x = gender_count.index, height = gender_count.values, width = 0.5)

plt.show()
```

```python
# 가로 막대 그래프(horizontal bar graph)
plt.barh(y = gender_count.index, width = gender_count.values, height= 0.5)

plt.show()
```

### Ex 4. 성별 tip의 box plot을 하나의 subplot에 그리기.

* x축: 남성/여성
* y축: 남성 box plot, 여성 box plot

```python
plt.boxplot(
    x = [tips[tips.sex == 'Male'].tip, tips[tips.sex == 'Female'].tip],
    labels = ['Male', 'Female']
)

plt.grid()

plt.show()
```

```python
sns.boxplot(
    x = 'sex',
    y = 'tip',
    data = tips,
    hue = 'smoker',
    palette = 'Set3'
)
# hue = 'smoker'
# hue 옵션을 사용해서 카테고리별 비교가 가능.
```

### Ex 5. 시간별 tip의 box plot을 하나의 subplot에 그리기.

```python
time_tip = []
for x in tips.time.unique():
    time_tip.append(tips[tips.time == x].tip)
time_tip
```

```python
plt.boxplot(
    x = time_tip,
    labels = tips.time.unique()
)

plt.grid()

plt.show()
```

```python
sns.boxplot(
    x = 'time',
    y = 'tip',
    data = tips,
    palette = 'Set3'
)
```
